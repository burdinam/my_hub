```python
#Load the required libraries

import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
#Loading data
```

```python
df = pd.read_csv("C:/Users/Ark/0_ANALYSIS/DA_projects/hotel_bookings/hotel_bookings.csv", sep=';')
```

```python
df.info()
```

```python
# Reduce column names to lowercase and replace spaces with underscores for ease of use.
```

```python
df.columns = [w.strip().lower().replace(' ', '_') for w in df.columns]
```

```python
df.columns
```

```python
df.head()
```

```python
# Top 5 countries by successful bookings
```

```python
top_5 = df\
    .query("is_canceled == 0")\
    .groupby("country", as_index=False)\
    .agg({"hotel":"count"})\
    .sort_values("hotel", ascending=False)\
    .rename(columns={"hotel":"total_bookings"})\
    .head()
```

```python
top_5
```

```python
# average number of booking nights
```

```python
avg_nights = df\
    .query("is_canceled == 0")\
    .groupby("hotel", as_index=False)\
    .agg({"stays_total_nights":"mean"})\
    .sort_values("stays_total_nights", ascending=False)    
```

```python
avg_nights
```

```python
# assigned_room_type and reserved_room_type difference/
```

```python
df.columns
```

```python
df[['reserved_room_type', 'assigned_room_type']]
```

```python
num_difference = df\
    .query("reserved_room_type != assigned_room_type")\
    .groupby('hotel', as_index = False)\
    .agg({"assigned_room_type":"count"})\
    .sort_values('assigned_room_type', ascending = False)

num_difference
```

```python
#popular month for booking
```

```python
top_months = df\
    .groupby(['arrival_date_month'], as_index = False)\
    .agg({"stays_total_nights":"sum"})\
    .sort_values("stays_total_nights", ascending=False)

top_months.head()
```

```python
plt.figure(figsize=(12, 8))
ax = sns.countplot(x=df["arrival_date_month"])
```

```python
top_3_months_yearly_hotels = df\
    .groupby(['hotel', 'arrival_date_year', 'arrival_date_month'], as_index = False)\
    .agg({"stays_total_nights":"sum"})\
    .sort_values(["hotel", "arrival_date_year", 'stays_total_nights'], ascending=[False, True, False])\
    .groupby(['hotel', "arrival_date_year"])\
    .head(3)

top_3_months_yearly_hotels
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x="arrival_date_month", hue="hotel")
```

```python
df.columns
```

```python
top_3_months_yearly_hotels_by_cancelation = df\
    .query("is_canceled != 0")\
    .groupby(['hotel', 'arrival_date_year', 'arrival_date_month'], as_index = False)\
    .agg({"stays_total_nights":"sum"})\
    .sort_values(["hotel", "arrival_date_year", 'stays_total_nights'], ascending=[False, True, False])\
    .groupby(['hotel', "arrival_date_year"])\
    .head(3)

top_3_months_yearly_hotels_by_cancelation
```
