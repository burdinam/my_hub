```python
ads_data.head()
```

```python
df = ads_data\
    .groupby("Date", as_index=False)["Small Bags", "Large Bags", "XLarge Bags"]\
    .agg({"Small Bags":"sum", "Large Bags":"sum", "XLarge Bags":"sum"})\
    .sort_values("Date", ascending = True)\
    .rename(columns={"Small Bags":"Small_Bags", "Large Bags":"Large_Bags", "XLarge Bags":"X_Large_Bags"})
```

```python
df.info()
```

```python
df.head()
```

# Matlab-style

## Rotation

```python
plt.style.use('ggplot')
plt.plot(df.Date, df.Large_Bags, color='k')
plt.plot(df.Date, df.X_Large_Bags, color='g')
plt.xticks(rotation=90)
```

## Legends

```python
plt.plot(df.Date, df.Large_Bags, color='k')
plt.plot(df.Date, df.Small_Bags, color='y')
plt.xticks(rotation=90)

plt.legend(["Large_Bags", "Small_Bags"], 
           loc ="upper left")
```

## All-in-one

```python
plt.style.use('ggplot')
plt.figure(figsize=(10, 5))

plt.plot(df.Date, df.Small_Bags, color='y', label = "Small_Bags")
plt.plot(df.Date, df.Large_Bags, color='k', label = "Large_Bags")

plt.xticks(fontsize=15, rotation=90)
plt.yticks(fontsize=10, rotation=45)

plt.xlim(left=17000, right=17500)
plt.ylim(bottom=min(df.X_Large_Bags)*2, top=max(df.Large_Bags)/2)

plt.xlabel('Date', fontsize=15)
plt.ylabel('Revenue', fontsize=15)
plt.title("Avokados' revenue", fontsize=25)

plt.legend(loc =(1.016, 0.0), borderaxespad=0, fontsize=20)

plt.savefig("C:/Users/Qwerty/ANALYSIS/my_hub/SEABORN/pic_1", dpi=300)
```
