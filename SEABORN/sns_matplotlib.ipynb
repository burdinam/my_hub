```python
games = pd.read_csv("C:/Users/Qwerty/ANALYSIS/my_hub/STATISTICS/data_folder/games.csv")
```

```python
games.head()
```

```python
count_games = games\
    .groupby("Year")\
    .agg({"Platform":"count", "Global_Sales":"sum"})\
    .rename(columns={"Platform":"Number_games"})\
    .sort_values("Year", ascending=True)
```

```python
count_games.head()
```

```python
g = (sns.jointplot(x="Number_games",
             y="Global_Sales",
             kind='scatter',
             data=count_games,
             marginal_kws=dict(bins=10, rug=True))
    .plot_joint(sns.kdeplot))


plt.show()
plt.clf()
```

```python
sns.lmplot(x='Number_games', y='Global_Sales', data=count_games, lowess=True, 
           aspect=2, line_kws={'color' : 'black'})
```

```python
# Create the same PairGrid but map a histogram on the diag
plt.style.use("ggplot")
g = sns.PairGrid(count_games, vars=["Number_games", "Global_Sales"])
g2 = g.map_diag(plt.hist)
g3 = g2.map_offdiag(plt.scatter)
plt.show()
plt.clf()
```
