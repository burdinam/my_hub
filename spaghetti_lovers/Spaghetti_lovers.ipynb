# Analysis

```python
user_df = df[['tc', 'art_sp']]
```

```python
#Renaming columns for better use
user_df = user_df.rename(columns={'tc':'user_id', 'art_sp':'brand_info'})
```

```python
#separating the brand name into another column
user_df['brand_name'] = user_df.brand_info.apply(lambda x: x.split(' ')[-1])
```

```python
user_df
```

```python
# Suppose there are loyal clients that buy regularly
user_purchases = user_df.groupby('user_id', as_index = False)\
                        .agg({'brand_info':'count'})\
                        .rename(columns={'brand_info':'num_purchases'})
```

```python
user_purchases
```

```python
user_purchases.describe()
```

```python
# 25% bought more than 5 times.
# Consider them as the most loyal clients.
```

```python
user_purchases = user_purchases.query("num_purchases >= 5").sort_values(['num_purchases'], ascending=False)
```

```python
user_purchases
```

```python
# let's find out the most favorite (top 5) brands by number of purchases
favorite_brands = user_df.groupby('brand_name', as_index = False)\
                        .agg({'brand_info':'count'})\
                        .rename(columns={'brand_info':'num_purchases'})\
                        .sort_values('num_purchases', ascending=False)\
                        .head()
```

```python
favorite_brands
```

```python
# Who are the clients who buy the most and which brand?
```

```python
user_df.head()
```

```python
spaghetti_lovers = user_df.groupby(['user_id', 'brand_name'], as_index = False)\
    .agg({'brand_info':'count'})\
    .sort_values(['user_id', 'brand_info'], ascending=[False, False])\
    .groupby('user_id')\
    .head(1)\
    .rename(columns={'brand_name': 'lovely_brand','brand_info': 'lovely_brand_purchases'})
```

```python
# top 10 spaghetti lovers
spaghetti_lovers.head(10)
```

```python
# Number of unique brands that people buy.
```

```python
unique_brands = user_df.groupby('user_id', as_index = False)\
    .agg({"brand_name" : pd.Series.nunique})\
    .rename(columns = {"brand_name":"num_unique_brands"})
```

```python
unique_brands.sort_values('num_unique_brands', ascending=False)
```

```python
#Gathering data
complete_df = user_purchases.merge(unique_brands, on='user_id')\
    .merge(spaghetti_lovers, on='user_id')
```

```python
complete_df.head()
```

```python
# Now, we can see there are 100% brand loyal clients (true pasta lovers) 
# and people that buy different brands but prefer their favorite.
```

```python
#---
```

```python
# Lets build a loyalty score indicator for clients
```

```python
complete_df['loyalty_score'] = complete_df.lovely_brand_purchases / complete_df.num_purchases
```

```python
complete_df
```

```python
# Visualising data
ax = sns.distplot(complete_df.loyalty_score, kde=False)
```

```python
# Median loyalty score for each brand
```

```python
brand_loyalty = complete_df \
    .groupby('lovely_brand', as_index = False)\
    .agg({"loyalty_score":"median", "user_id":"count"})\
    .sort_values("loyalty_score", ascending=False)
```

```python
brand_loyalty
```

```python
# loyalty by loyalty_score
sns.barplot(data=brand_loyalty, x="lovely_brand", y="loyalty_score")
```

```python
# loyalty by user choice
sns.barplot(data=brand_loyalty, x="lovely_brand", y="user_id")
```
