```python
# cohort analysis
```

```python
df = merged_df.copy()
```

```python
df.head()
```

```python
df = df[~df.order_status.isin(['created'])]
```

```python
df.dtypes
```

```python
df['month'] = df.order_purchase_timestamp.astype(str).apply(lambda x: x[:7])
```

```python
df.head()
```

```python
df = df.merge(
    df
    .groupby(['customer_unique_id'], as_index=False)
    .month.min(),
    on='customer_unique_id',
    how='left'
)
df.head()
```

```python
df['ret_month'] = (
    df
        .groupby('customer_unique_id', as_index=False)
        .month_y
        .cumcount()
)
```

```python
df['retention'] = 1
df.head()
```

```python
# Cohort
```

```python
df = pd.pivot_table(
        df,
        values = 'retention',
        index = 'month_y',
        columns = 'ret_month',
        aggfunc=np.sum
)
df
```

```python
df = df.div(df[0], axis=0).fillna(0).drop(0, axis=1)
df
```

```python
plt.figure(figsize=(20, 14))
sns.heatmap(df, robust=True, annot=True, cmap='YlGnBu')
```

```python
# # RFM analysis (clustering)
```

```python
# 1 выбрать методику расчета RFM-кластеров
# 2 Выбираем простейшую для e-comm
#     R время в днях с последнего заказа
#     F Количество заказов за все время
#     М общая сумма покупок за все время
# 3 Кластеризация простейшая. Разбивка по квартилям этих метрик
# 4 Расчитал время от каждого заказа до максимального возможного времени внутри датасэта
# 5 рассчитал стоимость заказа для каждого заказа в данных 
# 6 приклеил стоимости и уникальные ID пользователей к данным 
# 7 для каждого пользователя вычислил:
#     срок в днях от его последнего заказа
#     суммарное количество заказов 
#     общую сумму заказов 
# 8 с помощью np.digitize присвоил каждому пользователю квартиль внутри метрики    
```

```python
orders_ = pd.read_csv("C:/Users/Qwerty/ANALYSIS/my_hub/RFM/olist_orders_dataset.csv")
```

```python
items.head(1)
```

```python
customers.head(1)
```

```python
max_date = parser.parse(orders_.order_purchase_timestamp.max())
max_date
```

```python
orders_['days_to_max'] = list(map(lambda x: x.days, (max_date - orders_.order_purchase_timestamp.apply(parser.parse))))
```

```python
orders_.head()
```

```python
rfm_data = (
    orders_[~orders_.order_status.isin(['created'])]
    .merge(
        customers[["customer_id", "customer_unique_id"]],
        on="customer_id",
        how="left"
    )
    .merge(
        items
        .groupby(["order_id"], as_index=False)
        .price.sum()
    ))
    
rfm_data.head()
```

```python
rfm_data = rfm_data.groupby(['customer_unique_id'], as_index=False)\
    .agg({"order_id":"size", "days_to_max":"min", "price":"sum"})
rfm_data
```

```python
#Recency
```

```python
intervals_r = list(rfm_data.days_to_max.describe()[3:])
intervals_r
```

```python
#Frequency. 
intervals_f = [1, 2, 3, 16]
```

```python
intervals_m = list(rfm_data.price.describe())[3:]
intervals_m
```

```python
rfm_data['recency'] = rfm_data["days_to_max"].apply(lambda x: np.digitize(x, intervals_r))
rfm_data['frequency'] = rfm_data["order_id"].apply(lambda x: np.digitize(x, intervals_f))
rfm_data['monetary'] = rfm_data["price"].apply(lambda x: np.digitize(x, intervals_m))
rfm_data['rfm_rank'] = rfm_data['recency'].astype(str) + rfm_data['frequency'].astype(str) + rfm_data['monetary'].astype(str)
```

```python
rfm_data
```

```python
rfm_data.groupby(["rfm_rank"], as_index=False)["order_id", "days_to_max", "price"].mean()
```
