```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
london = pd.read_csv("C:/Users/Qwerty/ANALYSIS/my_hub/STATISTICS/data_folder/london.csv", parse_dates=['timestamp'])
```

```python
london.isna().sum()
```

```python
london.dtypes
```

```python
london = london.astype({'weather_code':'int64', 'is_holiday':'int64', 'is_weekend':'int64', 'season':'int64'})
```

```python
# Number of rides
```

```python
ax = sns.relplot(data=london, x = 'timestamp', y="cnt", kind="line")
ax.set_xticklabels(rotation=45)
plt.show()
```

```python
# Rides by days
```

```python
london\
    .set_index(pd.DatetimeIndex(london['timestamp']))\
    .resample(rule='D')\
    .agg({'cnt' : 'sum'})\
    .sort_values('timestamp', ascending = True)\
    .plot()

plt.show()
```

```python
# Moving average
```

```python
MA = london\
    .set_index(pd.DatetimeIndex(london['timestamp']))\
    .resample(rule='D')\
    .agg({'cnt' : 'sum'})\
    .sort_values('timestamp', ascending = True)    
    
MA.head()
# plt.show()
```

```python
MA.rolling(window=10).mean().plot()
plt.show()
```

```python
MA = MA.assign(MovAv = MA.rolling(window=10).mean())
MA = MA.assign(difference = MA.MovAv - MA.cnt)
```
