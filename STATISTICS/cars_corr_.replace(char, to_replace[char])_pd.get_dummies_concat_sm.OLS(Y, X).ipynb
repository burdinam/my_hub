```python
# separating the company name
```

```python
cars["company"] = cars.CarName.apply(lambda x: x.split(' ')[0])
```

```python
cars.company.unique()
```

```python
# formating company's name
```

```python
cars["company"] = cars.company.apply(lambda x: x.lower())
```

```python
to_replace = {"maxda": "mazda", "porcshce": "porsche", "toyouta": "toyota", "vokswagen": "volkswagen", "vw": "volkswagen"}

# .keys() returns a list of all the keys in the dictionary
for char in to_replace.keys():
    # dictionaryName[keys] - another way to get value
    cars["company"] = cars.company.replace(char, to_replace[char])
```

```python
df = cars.loc[:, ['price', 'company', 'fueltype', 'aspiration','carbody', 'drivewheel', 'wheelbase', 'carlength',\
    'carwidth', 'curbweight', 'enginetype', 'cylindernumber', 'enginesize', 'boreratio','horsepower']]
```

```python
# add pivoted columns == "objects"
```

```python
df_dummy = pd.get_dummies(data=cars[["company", 'fueltype', 'aspiration', 'carbody', 'drivewheel', 'enginetype',
       'cylindernumber']], drop_first = True)
```

```python
df_dummy.head()
```

```python
# join together
```

```python
dff = pd.concat([df, df_dummy], axis=1)
```

```python
dff.drop(['company', 'fueltype', 'aspiration','carbody', 'drivewheel', \
                'enginetype', 'cylindernumber'], axis='columns', inplace=True)
```

```python
# delete columns with unnecessary data
```

```python
dff.head()
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf 
```

```python
Y = dff.price #dependent variable
X = sm.add_constant(dff.drop(columns='price'))  # all dependent variables (excl: price)
model = sm.OLS(Y, X)  
results = model.fit()  # build regression
print(results.summary())
```
